## Which of the Indian online retailer would you recommend to a friend?

```python
df['Which of the Indian online retailer would you recommend to a friend?'].value_counts()
```

```python
df1=df.copy()
```

```python

def retValues(x):
    am,fl,pa,my,sn=0,0,0,0,0
    #print(x)
    if "amazon" in x.lower():
        am=1
    if "flipkart" in x.lower():
        fl=1
    if "paytm" in x.lower():
        pa=1
    if "myntra" in x.lower():
        my=1    
    if "snapdeal" in x.lower():
        sn=1
    #print(am,fl,pa,my,sn)
    FlipArray.append(fl)
    AmArray.append(am)
    paytmArray.append(pa)
    myntArray.append(my)
    snapArray.append(sn)
```

```python
FlipArray=[]
AmArray=[]
paytmArray=[]
myntArray=[]
snapArray=[]
df['Which of the Indian online retailer would you recommend to a friend?'].apply(retValues);
```

```python
df['FlipKartArray']=FlipArray
df['AmazonArray']=AmArray
df['PaytmArray']=paytmArray
df['MyntraArray']=myntArray                                       
df['SnapArray']=snapArray
```
