```python
driver2=webdriver.Chrome(r"E:\softwares\chromedriver.exe",options=option)
```

```python
def getCarDetails(driver2,atag):
    flag=1
    driver2.get(atag)   
    #scroll_to_end(driver2)
    driver2.implicitly_wait(1)
    dictSpecs=dict()
    year=' '
    kms=' '
    owner=' '
    vehicleType=' '
    manual=' '
    location=' '
    amount=' '
    brandName=' '
    try:
        name=driver2.find_element_by_xpath(".//div[@class='vdp-head']/h1[2]").text
        brandName=name.split()[0]

        location='Hyderabad'
        amount=driver2.find_element_by_xpath(".//span[@class='amount']").text
        #data-read-more="1"
        try:
            GenDetailsCon=driver2.find_element_by_xpath(".//div[@class='GenDetailBox']")
            liItems=GenDetailsCon.find_elements_by_xpath(".//li[@class='gsc_col-xs-6']")
            year=liItems[0].find_element_by_xpath(".//div/div").text
            kms=liItems[2].find_element_by_xpath(".//div/div").text
            owner=liItems[3].find_element_by_xpath(".//div/div").text
            vehicleType=liItems[4].find_element_by_xpath(".//div/div").text
            manual=liItems[5].find_element_by_xpath(".//div/div").text


        except NoSuchElementException:
            year=''
            kms=''
            owner=''
            vehicleType=''
            manual=''

        
        try:
            search_btn=driver2.find_element_by_xpath("//a[@class=' viewMorebtn viewMorebtnArrow']")
            search_btn.click()
            specs=driver2.find_element_by_xpath(".//div[@class='usedSpecFeatureBoxs ']")

            liItems=specs.find_elements_by_xpath(".//li[@class='gsc_col-xs-6']")
            for liItem in liItems:
                typeItem=liItem.find_element_by_xpath(".//div[@class='smallSpec']").text
                typeVal=liItem.find_element_by_xpath(".//div[@class='largeSpec']").text
                dictSpecs[typeItem]=typeVal
        except:
            pass
    except:
        flag=0
    
    return(flag,year,kms,owner,vehicleType,manual,location,amount,brandName,dictSpecs)
```

```python
yearLst=[]
kmsLst=[]
ownerLst=[]
vehicleTypeLst=[]
manualLst=[]
locationLst=[]
amountLst=[]
brandNameLst=[]

mileageLst=[]
engineLst=[]
maxPowerLst=[]
maxPowerRpmLst=[]
seatsLst=[]
ColorLst=[]
GearBoxLst=[]
driveTypeLst=[]
steeringTypeLst=[]
FrontBrkTypeLst=[]
RearBrkTypeLst=[]
TopSpeedLst=[]
accelarationLst=[]
DoorsLst=[]
maxTorqueLst=[]
maxTorqueRpmLst=[]
no_CylindersLst=[]
ValvesCylinderLst=[]
valveConfigLst=[]
fuelSupplyLst=[]



for divContainer in divContainers:
    atag= divContainer.find_element_by_xpath(".//div[@class='gsc_col-xs-7  carsName']/a").get_attribute('href')
    (flag,year,kms,owner,vehicleType,manual,location,amount,brandName,dictSpecs)=getCarDetails(driver2,atag)
    
    
    if flag==1:
        yearLst.append(year)
        kmsLst.append(kms)
        ownerLst.append(owner)
        vehicleTypeLst.append(vehicleType)
        manualLst.append(manual)
        locationLst.append(location)
        amountLst.append(amount)
        brandNameLst.append(brandName)



        mileageLst.append(dictSpecs.get('Mileage',''))
        engineLst.append(dictSpecs.get('Engine',''))
        maxpowerRPM=dictSpecs.get('Max Power','')
        maxpower=''
        maxpowerRpm=''
        try:
            maxpower=maxpowerRPM.split('@')[0]
            maxpowerRpm=maxpowerRPM.split('@')[1]
            maxPowerLst.append(maxpower)
            maxPowerRpmLst.append(maxpowerRpm)
        except:
            maxPowerLst.append(maxpower)
            maxPowerRpmLst.append(maxpowerRpm)
        seatsLst.append(dictSpecs.get('Seats',''))
        ColorLst.append(dictSpecs.get('Color',''))
        GearBoxLst.append(dictSpecs.get('Gear Box',''))
        driveTypeLst.append(dictSpecs.get('Drive Type',''))
        steeringTypeLst.append(dictSpecs.get('Steering Type',''))
        FrontBrkTypeLst.append(dictSpecs.get('Front Brake Type',''))
        RearBrkTypeLst.append(dictSpecs.get('Rear Brake Type',''))
        TopSpeedLst.append(dictSpecs.get('Top Speed',''))
        accelarationLst.append(dictSpecs.get('Acceleration',''))
        DoorsLst.append(dictSpecs.get('No Of Doors',''))




        maxTorqueRPM=maxTorqueLst.append(dictSpecs.get('Max Torque',''))
        maxTorque=''
        maxTorqueRpm=''
        try:
            maxTorque=maxTorqueRPM.split('@')[0]
            maxTorqueRpm=maxTorqueRPM.split('@')[1]
            maxTorqueLst.append(maxTorque)
            maxTorqueRpmLst.append(maxTorqueRpm)
        except:
            maxTorqueLst.append(maxTorque)
            maxTorqueRpmLst.append(maxTorqueRpm)
            
        no_CylindersLst.append(dictSpecs.get('No Of Cylinder',''))
        ValvesCylinderLst.append(dictSpecs.get('Valves Per Cylinder',''))
        valveConfigLst.append(dictSpecs.get('Valve Configuration',''))
        fuelSupplyLst.append(dictSpecs.get('Fuel Supply System',''))

```

```python
pd.set_option('display.max_columns',None)
```

```python
carsDFHyd=pd.DataFrame()
carsDFHyd['year']= yearLst

carsDFHyd['kms']=kmsLst 
carsDFHyd['owners']=ownerLst 
carsDFHyd['vehicle']=vehicleTypeLst 
carsDFHyd['manual']=manualLst 
carsDFHyd['location']=locationLst 
carsDFHyd['amount']=amountLst 
carsDFHyd['brand']=brandNameLst 

carsDFHyd['mileage']=mileageLst 
carsDFHyd['engine']=engineLst 
carsDFHyd['maxPower']=maxPowerLst 
carsDFHyd['maxPowerRPM']=maxPowerRpmLst 
carsDFHyd['seats']=seatsLst 
carsDFHyd['Color']=ColorLst 
carsDFHyd['GearBox']=GearBoxLst 
carsDFHyd['driveType']=driveTypeLst 
carsDFHyd['SteeringType']=steeringTypeLst 
carsDFHyd['FrontBrkType']=FrontBrkTypeLst 
carsDFHyd['RearBrkType']=RearBrkTypeLst 
carsDFHyd['TopSpeed']=TopSpeedLst 
carsDFHyd['acceleration']=accelarationLst 

carsDFHyd['Doors']=DoorsLst
#carsDFHyd['maxTorque'] =maxTorqueLst
#carsDFHyd['maxTorqueRPM']=maxTorqueRpmLst
carsDFHyd['no_cyclinders'] =no_CylindersLst
carsDFHyd['no_valves'] =ValvesCylinderLst
carsDFHyd['valveConfig'] =valveConfigLst
carsDFHyd['fuelSupply'] =fuelSupplyLst




carsDFHyd['index']=list(range(1,len(brandNameLst)+1))
carsDFHyd.set_index('index',inplace=True)

carsDFHyd
    
```
