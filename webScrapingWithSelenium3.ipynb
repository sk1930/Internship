### Scrape the details of most viewed videos on YouTube from Wikipedia:
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/
Rank, Name, Artist,Upload date, views

```python
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
tableContainer=driver.find_element_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']")
tableContainer=tableContainer.find_elements_by_xpath(".//tbody/tr")
```

```python
def getVideoDetails(tableRoww):
    try:
        tdItems=tableRoww.find_elements_by_xpath(".//td")
        rank=tdItems[0].text.strip(".")

        videoName=tdItems[1].text
        artist=tdItems[2].text
        views=tdItems[3].text
        uploadDat=tdItems[4].text


        

    except NoSuchElementException:
        rank="_"
        videoName="_"
        artist="_"
        views="_"
        uploadDat="_"
   
       
    
    return(rank,videoName,artist,views,uploadDat)
```

```python


rankLst=[]
nameLst=[]
artistLst=[]
viewLst=[]
uploadDateLst=[]

for tableRoww in tableContainer:
    (rank,videoName,artist,views,uploadDat)=getVideoDetails(tableRoww)

    rankLst.append(rank)
    nameLst.append(videoName)
    artistLst.append(artist)
    viewLst.append(views)
    uploadDateLst.append(uploadDat)
```

```python

videosDF=pd.DataFrame()
videosDF['Rank']=rankLst
videosDF['video Name']=nameLst
videosDF['Artist']=artistLst
videosDF['views']=viewLst
videosDF['uploadDate']=uploadDateLst



videosDF['index']=list(range(1,len(rankLst)+1))
videosDF.set_index('index',inplace=True)
videosDF
    
```

```python
def getSeriesDetails(listItem):
    try:
        dateTimeCon=listItem.find_element_by_xpath(".//div[@class='fixture__datetime desktop-only']")
        day=dateTimeCon.find_element_by_xpath(".//span[@class='fixture__day']")
        day=day.text
        month=dateTimeCon.find_element_by_xpath(".//span[@class='fixture__month']")
        month=month.text
        date=dateTimeCon.find_element_by_xpath(".//span[@class='fixture__date']")
        date=date.text
        dateDay=day+" "+date+" " +month

    except:
        dateDay='-'       
    try:
        time=dateTimeCon.find_element_by_xpath(".//span[@class='fixture__time']")
        time=time.text

    except NoSuchElementException:
        time="-"
    try:
       
        Series=listItem.find_element_by_xpath(".//div[@class='fixture__teams']")
        Series=Series.text.replace("\n"," ")

    except:
        Series="-"
    try:
        matchPlace=listItem.find_element_by_xpath(".//p[@class='fixture__additional-info']")
        matchPlace=matchPlace.text.split("\n")
        match=matchPlace[0]
        place=matchPlace[1]

    except:
        match="-"
        place="_"          
    
    return(dateDay,time,Series,match,place)
```

```python

dateDayLst=[]
timeLst=[]
seriesLst=[]
matchLst=[]
placeLst=[]

listConta=driver.find_elements_by_xpath("//a[contains(@class,'js-match event-list__item')]")
for listItem in listConta:
    (dateDay,time,Series,match,place)=getSeriesDetails(listItem)

    dateDayLst.append(dateDay)
    timeLst.append(time)
    seriesLst.append(Series)
    matchLst.append(match)
    placeLst.append(place)
```

```python

SeriesDF=pd.DataFrame()
SeriesDF['Match Title']=matchLst
SeriesDF['place']=placeLst
SeriesDF['date']=dateDayLst
SeriesDF['Time']=timeLst
SeriesDF['Series']=seriesLst



SeriesDF['index']=list(range(1,len(matchLst)+1))
SeriesDF.set_index('index',inplace=True)
SeriesDF
    
```
